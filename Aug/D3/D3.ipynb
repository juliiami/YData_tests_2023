```python
def solution(inp):
    hs = list(map(int, inp.split()))
    sol1 = []
    mx = 0
    for h in hs:
        mx = max(mx, h)
        sol1.append(mx - h)
    sol2 = []
    mn = hs[-1]
    for h in hs[::-1]:
        mn = min(mn, h)
        sol2.append(h - mn)
    result = 0
    best = 0
    sol2.reverse()
    for i in range(len(hs)):
        result = max(result, best + sol2[i])
        best = max(best, sol1[i])
    return(result)
```
