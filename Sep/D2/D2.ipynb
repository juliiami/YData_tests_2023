```python
def solution(x, A, B):
    A = [int(i) for i in A.split()]
    B = [int(i) for i in B.split()]
    n = len(A)
    m = len(B) 
    s = set(A)
    count=0
    for i in range (0, n):
        s.add(A[i])
    for j in range(0, m):
        if ((x - B[j]) in s):
            count+=1
    return count
```
