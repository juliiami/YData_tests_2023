```python
import sys

def solution(x, A, B):
    A = [int(i) for i in A.split()]
    B = [int(i) for i in B.split()]
    m = len(A)
    n = len(B) 
    A.sort()
    B.sort()
    diff=sys.maxsize

    l = 0
    r = n-1
    res_l, res_r = 0, 0
    while(l < m and r >= 0):
        if abs(A[l] + B[r] - x) < diff:
            res_l = l
            res_r = r
            diff = abs(A[l] + B[r] - x)
        if abs(A[l] + B[r] - x) == diff and (A[res_l], B[res_r]) > (A[l], B[r]):
            res_l = l
            res_r = r

            
        if A[l] + B[r] > x:
            r=r-1
        elif A[l] + B[r] < x:
            l=l+1
        else:
            break


    return([A[res_l], B[res_r]])
```
