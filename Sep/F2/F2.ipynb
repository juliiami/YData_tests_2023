```python
import pandas as pd
import numpy as np

stream_data = pd.read_csv('/tmp/deps/stream_data.csv', sep=",")

def solution(Language):
    stream_data['Growth'] = stream_data['Followers gained']/stream_data['Followers total']
    idx = stream_data[stream_data.Language==Language].Growth.idxmax()
    return(stream_data.ChannelID[idx])
```
